###### 1. BUILDING CONTENT BASED RECOMMENDER SYSTEM CONSIDERING THE HYPOTHETICAL DATASET CONTAINING 8 DOCUMENTS 
######    AND 10 KEYWORDS WITH 3 POSSIBLE RATINGS (0,1,2), WITH THEIR FREQUENCIES AS 0,1 AND 2, THEREFORE, 
######    DETERMINING THE 5 MOST DISCRIMINATIVE KEYWORDS WITH THE HELP OF GINI INDEX MEASURE, AND PRINTING 
######    THE FINAL TERM-FREQUENCY DOCUMENT CORRESPONDING TO THE DISCRIMINATIVE KEYWORDS.

```python
#### IMPORTING LIBRARIES
import numpy as np
```

```python
n_docs = 8  ## NO. OF DOCUMENTS/ROWS
n_words = 10 ## NO. OF KEYWORDS/COLUMNS
n_rts = 3   ## NO. OF POSSIBLE RATINGS
k = n_docs * n_words ## NO. OF TOTAL ELEMENTS IN THE 8*10 MATRIX
n_freqs = 3   ## NO. OF FREQUENCIES/TIMES THE WORDS OCCUR IN DOCUMENTS
mat_1d = np.random.choice(n_freqs, k)
mat_2d = mat_1d.reshape(n_docs, n_words)
mat_2d
```

```python
### DEFINING THE RATINGS CORRESPONDING TO EACH OF THE 8 DOCUMENTS IN THE MATRIX
mat_ratings = np.random.choice(n_rts, n_docs)
mat_ratings
```

```python
gi = np.zeros(n_words)
print("1_D ARRAY FOR STORING GINI INDEX VALUES FOR 10 KEYWORDS:", gi)
```

```python
### CREATING A LOOP FOR EVERY KEYWORD DISPLAYING RATED DOCUMENTS

for j in range(n_words):   ## DEFINING THE COLUMNS/KEYWORDS
    doc_list = []  ## STORING THE INDICES OF THE jth KEYWORD IN ith DOCUMENT

   ### FINDING THE NO. OF DOCUMENTS WHERE THE KEYWORD OCCURS (NON-ZER0 FREQUENCY)
    for i in range(n_docs):
        if mat_2d[i, j] != 0:  
            doc_list.append(i)
    
  
    print(f"\nKEYWORD {j}:")
    print("\nDOCUMENTS WHICH HAVE RECEIVED RATINGS FOR THIS KEYWORD:", doc_list)
    print("\nTHE RATINGS CORRESPONDING TO THE DOCUMENTS ARE:", mat_ratings[doc_list])


    n, c = np.unique(mat_ratings[doc_list], return_counts=True)
    print("\nTHE NUMBER OF TIMES EACH RATING OCCURS ARE:", c)

    
    probability = c / sum(c)
    print("\nTHE PROBABILITIES FOR EACH RATING ARE:", probability)

   ## COMPUTING THE GINI INDEX FOR EACH KEYWORD
    gi_value = 1 - np.sum(probability**2)
    gi[j] = gi_value
    print(f"\nGINI INDEX FOR KEYWORD {j}: {gi_value}")

## PRINTING THE GINI INDICES FOR EACH KEYWORDS
print("\nGINI INDEX VALUES FOR EACH KEYWORD:")
print(gi)

## SORTING THE GINI INDICES
sorted_gi = np.sort(gi)
print("\nSORTED GINI INDEX VALUES:")
print(sorted_gi)

    
```

```python
### LISTING OF THE 5 MOST DISCRIMINATIVE KEYWORDS
most_discriminative_keywords = np.argsort(gi)[:5]
print("\nMOST DISCRIMINATIVE KEYWORDS (BASED ON GINI INDEX VALUE):", most_discriminative_keywords)
```

```python
### PRINTING TH FINAL TFD CORRESPONDING TO THE DISCRIMINATIVE WORDS
final_tfd = mat_2d[:, most_discriminative_keywords]
print("THE FINAL TERM FREQUENCY DOCUMENT IS:\n", final_tfd)
```

###### 2. BUILDING CONTENT BASED RECOMMENDER SYSTEM CONSIDERING THE HYPOTHETICAL DATASET CONTAINING 25 DOCUMENTS
######    AND 75 KEYWORDS WITH 4 POSSIBLE RATINGS (0,1,2,3), WITH THEIR FREQUENCIES AS 0,1,2,3, THEREFORE, DETERMINING
######    THE 10 MOST DISCRIMINATIVE KEYWORDS WITH THE HELP OF ENTROPY, AND PRINTING THE FINAL TERM-FREQUENCY DOCUMENT 
######    CORRESPONDING TO THE DISCRIMINATIVE KEYWORDS.

```python
n_docs = 25  ## NO. OF DOCUMENTS/ROWS
n_words = 75 ## NO. OF KEYWORDS/COLUMNS
n_rts = 4   ## NO. OF POSSIBLE RATINGS
k = n_docs * n_words ## NO. OF TOTAL ELEMENTS IN THE 25*75 MATRIX
n_freqs = 4   ## NO. OF FREQUENCIES/TIMES THE WORDS OCCUR IN DOCUMENTS
mat_1d = np.random.choice(n_freqs, k)
mat_2d = mat_1d.reshape(n_docs, n_words)
mat_2d
```

```python
### DEFINING THE RATINGS CORRESPONDING TO EACH OF THE 25 DOCUMENTS IN THE MATRIX
mat_ratings = np.random.choice(n_rts, n_docs)
mat_ratings
```

```python
en = np.zeros(n_words)
print("1_D ARRAY FOR STORING ENTROPY VALUES FOR 75 KEYWORDS:", en)
```

```python
for j in range(n_words):   ## DEFINING THE COLUMNS/KEYWORDS
    doc_list = []  ## STORING THE INDICES OF THE jth KEYWORD IN ith DOCUMENT

   ### FINDING THE NO. OF DOCUMENTS WHERE THE KEYWORD OCCURS (NON-ZER0 FREQUENCY)
    for i in range(n_docs):
        if mat_2d[i, j] != 0:  
            doc_list.append(i)
    
  
    print(f"\nKEYWORD {j}:")
    print("\nDOCUMENTS WHICH HAVE RECEIVED RATINGS FOR THIS KEYWORD:", doc_list)
    print("\nTHE RATINGS CORRESPONDING TO THE DOCUMENTS ARE:", mat_ratings[doc_list])


    n, c = np.unique(mat_ratings[doc_list], return_counts=True)
    print("\nTHE NUMBER OF TIMES EACH RATING OCCURS ARE:", c)

    
    probability = c / sum(c)
    print("\nTHE PROBABILITIES FOR EACH RATING ARE:", probability)

   ## COMPUTING THE ENTROPY FOR EACH KEYWORD
    entropy = -np.sum(probability * np.log(probability))
    en[j] = entropy
    print(f"\nENTROPY FOR KEYWORD {j}: {entropy}")

## PRINTING THE GINI INDICES FOR EACH KEYWORDS
print("\nENTROPY VALUES FOR EACH KEYWORD:", en)

## SORTING THE ENTROPIES
sorted_en = np.sort(en)
print("\nSORTED GINI INDEX VALUES:")
print(sorted_en)
```

```python
### LISTING OF THE 10 MOST DISCRIMINATIVE KEYWORDS
most_discriminative_keywords = np.argsort(en)[:10]
print("\nMOST DISCRIMINATIVE KEYWORDS (BASED ON ENTROPY VALUE):", most_discriminative_keywords)
```

```python
### PRINTING TH FINAL TFD CORRESPONDING TO THE DISCRIMINATIVE WORDS
final_tfd = mat_2d[:, most_discriminative_keywords]
print("THE FINAL TERM FREQUENCY DOCUMENT IS:\n", final_tfd)
```
